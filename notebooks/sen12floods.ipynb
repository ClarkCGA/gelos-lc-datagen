The SEN12-FLOOD Dataset is built from 336 time series containing Sentinel 1 and Sentinel 2 images of areas that suffered a major flooding event during the 2019 winter. The acquisition period goes from December 2018 to May 2019. The observed areas are clustered in East Africa, South West Africa, Middle-East and Australia. A sequence corresponds to several tiles of size 512x512 each of them corresponding to a crop of a given acquisition. The global characteristics are summed up in the following table.
